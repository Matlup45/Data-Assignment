```python
import pandas as pd
import json

# Load JSON data
with open('/content/nested_data.json', 'r') as file:
    data = json.load(file)
```

```python
# Extract the 'programs' key
programs_list = data.get('programs', [])

# Create lists to store normalized data
programs = []
concerts = []
works = []
soloists = []

# Extract the 'programs' key
programs_list = data.get('programs', [])
```

```python


# Process each entry in the list of programs
for program in programs_list:
    season = program.get('season', '')
    orchestra = program.get('orchestra', '')
    program_id = program.get('programID', '')

    # Process concerts
    for concert in program.get('concerts', []):
        concerts.append({
            'programID': program_id,
            'Date': concert.get('Date', ''),
            'eventType': concert.get('eventType', ''),
            'Venue': concert.get('Venue', ''),
            'Location': concert.get('Location', ''),
            'Time': concert.get('Time', '')
        })

    # Process works
    for work in program.get('works', []):
        work_id = work.get('ID', '')
        works.append({
            'ID': work_id,
            'workTitle': work.get('workTitle', ''),
            'composerName': work.get('composerName', ''),
            'conductorName': work.get('conductorName', ''),
            'movement': work.get('movement', ''),
            'programID': program_id  # Link works to the program
        })

        # Process soloists
        for soloist in work.get('soloists', []):
            soloists.append({
                'workID': work_id,
                'soloistName': soloist.get('soloistName', ''),
                'soloistRoles': soloist.get('soloistRoles', ''),
                'soloistInstrument': soloist.get('soloistInstrument', '')
            })



```

```python
# Convert lists to dataframes
df_programs = pd.DataFrame(programs)
df_programs
```

```python
df_concerts = pd.DataFrame(concerts)
df_concerts
```

```python
df_works = pd.DataFrame(works)
df_works
```

```python
df_soloists = pd.DataFrame(soloists)
df_soloists
```

```python
# Save to CSV
df_programs.to_csv('programs.csv', index=False)
df_concerts.to_csv('concerts.csv', index=False)
df_works.to_csv('works.csv', index=False)
df_soloists.to_csv('soloists.csv', index=False)
```
