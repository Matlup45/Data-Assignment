```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Merge datasets on applicant_id
merged_df = pd.merge(applicants_df, loans_df, on='applicant_id')

selected_columns = [
    'Primary_applicant_age_in_years',
    'Number_of_dependents',
    'Years_at_current_residence',
    'Foreign_worker',
    'Months_loan_taken_for',
    'Principal_loan_amount',
    'EMI_rate_in_percentage_of_disposable_income',
    'Number_of_existing_loans_at_this_bank',
    'high_risk_applicant'
]

# Create a subset dataframe
subset_df = merged_df[selected_columns]

# Optionally, take a sample of the data to speed up the pairplot generation
sampled_df = subset_df.sample(frac=0.1, random_state=42)

# Generate the pairplot with the sampled data and selected features
sns.pairplot(sampled_df, hue='high_risk_applicant')
plt.show()
```

# ***Segment customers based on their risk***

```python
merged_df = pd.merge(applicants_df, loans_df, on='applicant_id')

segments = merged_df.groupby(['Primary_applicant_age_in_years', 'Employment_status', 'Marital_status', 'high_risk_applicant']).size().unstack()
segments
```

```python
def plot_segmented_distribution(data, feature, target, figsize=(10, 6)):
    plt.figure(figsize=figsize)
    sns.countplot(y=feature, hue=target, data=data, palette='coolwarm', order=data[feature].value_counts().index)
    plt.title(f'Distribution of {feature} by {target}')
    plt.show()

categorical_vars = ['Gender', 'Marital_status', 'Housing', 'Employment_status', 'Purpose', 'Property', 'Loan_history']
for var in categorical_vars:
    plot_segmented_distribution(merged_df, var, 'high_risk_applicant')
```

```python
low_risk_segments = merged_df[merged_df['high_risk_applicant'] == 0]
high_risk_segments = merged_df[merged_df['high_risk_applicant'] == 1]

low_risk_segments.groupby(['Primary_applicant_age_in_years', 'Employment_status']).size().unstack()
high_risk_segments.groupby(['Primary_applicant_age_in_years', 'Employment_status']).size().unstack()
```

```python
# Numerical plots
numerical_vars = ['Primary_applicant_age_in_years', 'Number_of_dependents', 'Principal_loan_amount', 'EMI_rate_in_percentage_of_disposable_income', 'Number_of_existing_loans_at_this_bank']

for var in numerical_vars:
    plt.figure(figsize=(10, 6))
    sns.histplot(merged_df[var], kde=True, bins=30)
    plt.title(f'Distribution of {var}')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='high_risk_applicant', data=merged_df)
plt.title('Distribution of High Risk Applicants')
plt.show()

```
