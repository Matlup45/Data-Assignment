```python
import pandas as pd

# Load the data
file_path = '/content/Time Series.xlsx'
df = pd.read_excel(file_path)
```

```python
df.describe(include='all')
```

```python
from datetime import datetime

# Convert start and end times to datetime
df['start_time'] = pd.to_datetime(df['Start'])
df['end_time'] = pd.to_datetime(df['End'])

# Sort the dataframe by Name and start_time
df = df.sort_values(by=['Name', 'start_time'])
df
```

```python
# Function to merge intervals
def merge_intervals(intervals):
    merged = []
    for interval in intervals:
        if not merged or merged[-1][1] < interval[0]:
            merged.append(interval)
        else:
            merged[-1] = (merged[-1][0], max(merged[-1][1], interval[1]), merged[-1][2] + interval[2])
    return merged
```

```python
# Group by Name and apply the merge_intervals function
result = []
for name, group in df.groupby('Name'):
    intervals = list(zip(group['start_time'], group['end_time'], group['Activity'].apply(lambda x: [x])))
    merged_intervals = merge_intervals(intervals)
    for interval in merged_intervals:
        result.append([name, interval[0], interval[1], interval[2]])

# Create a result dataframe
result_df = pd.DataFrame(result, columns=['Name', 'start_time', 'end_time', 'activities'])

# Display the result
result_df
```

```python
result_df.to_excel('result.xlsx', index=False)
```
